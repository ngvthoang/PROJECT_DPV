### Group project - Final exam

**Project Nhóm: Sức mạnh của Data Preparation qua Data Storytelling**

**Đề bài:** Sử dụng các kỹ thuật data storytelling để chứng minh và khẳng định vai trò quan trọng của Data Preparation (chuẩn bị dữ liệu) trong một dự án Machine Learning (học máy).

---

#### Yêu cầu sản phẩm & Đánh giá

**Phần 1: Data Storytelling (30%)**
*   **Nội dung:** Xây dựng một câu chuyện dữ liệu (dạng slide/báo cáo) để thuyết phục.
*   **Yêu cầu:** Câu chuyện phải so sánh rõ ràng sự khác biệt (về insights, trực quan hóa, hoặc kết quả mô hình ML đơn giản) khi sử dụng dữ liệu "thô" so với dữ liệu đã qua chuẩn bị kỹ lưỡng (làm sạch, biến đổi, chọn lọc đặc trưng).

**Phần 2: Phân tích Kỹ thuật (40%)**
*   **Nội dung:** Viết báo cáo phân tích cách thức xây dựng Phần 1.
*   **Yêu cầu:** Giải thích rõ:
    *   Câu chuyện tuân theo chiến thuật, quy tắc kể chuyện nào? (Tham khảo *Storytelling with Data*).
    *   Các biểu đồ đã chọn tuân theo nguyên tắc trực quan hóa nào (ví dụ: giảm clutter, dùng thuộc tính tiền chú ý, lựa chọn biểu đồ...)?

**Phần 3: Code Project trên GitHub (30%)**
*   **Nội dung:** Một repository GitHub công khai chứa toàn bộ code (Python).
*   **Yêu cầu:**
    *   Code tổ chức khoa học, có chú thích đầy đủ.
    *   File `README.md` chuyên nghiệp: mô tả dự án, bộ dữ liệu, hướng dẫn chạy code.
